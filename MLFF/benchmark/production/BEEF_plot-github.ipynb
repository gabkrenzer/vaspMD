```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
T_list = [450, 500, 550, 600, 900]
n_steps = 501000
tstep = 0.002 #ps
```

```python
fs = 13
```

```python
dt = np.linspace(1, n_steps*tstep, n_steps, True)
```

```python
beef_master = []
```

```python
for T in T_list:
    beef_T = []
    df = pd.read_csv('BEEF'+str(T)+'.dat', names = ['max_force'])
    for i in range(0, n_steps):
        beef_T.append(df.loc[i,'max_force'])
    beef_master.append(beef_T)
```

```python
for i in range(0, len(T_list)):
    plt.title(str(T_list[i])+'K', fontsize = 15)
    plt.plot(dt, beef_master[i])
    plt.xticks(fontsize = fs)
    plt.yticks(fontsize = fs)
    plt.xlabel('Time [ps]', fontsize = fs)
    plt.ylabel('BEEF', fontsize = fs)
    plt.show()        
```

Here, we can see that the Bayesian errors stay constant throughout the production run. We can trust the data.
